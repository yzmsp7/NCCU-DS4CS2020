4-1. Do some data preprocessing and feature engineering. Please leave some comments in your codes. At last, please print the first 5 rows of your new dataframe (name it as "ca_my_df"). 


```
!pip install reverse_geocoder
```

```
import reverse_geocoder
def rgr_batch(df):
    cord = []
    for idx in df.index:
        cord.append((df['latitude'][idx], df['longitude'][idx]))
    result = reverse_geocoder.search(cord)
    return pd.DataFrame(result).drop(columns=['lat', 'lon'])
```

```
# make a backup
ca_my_df = ca_df.copy()

# Try to do data preprocessing and feature engineering here. save the result in 'ca_my_df'.

# 1. Because there's latitude and longitude data, we can find the area in this data.
# and just retain the county information
location = rgr_batch(ca_df[['latitude','longitude']])['admin2'].rename('county')
ca_my_df = pd.concat([ca_my_df, location], axis=1)
# one hot encoding
ca_my_df = pd.concat([ca_my_df, pd.get_dummies(ca_my_df.county)], axis=1).drop('county',axis=1)

# 2. we can calculate the average of households own their rooms and bedrooms
ca_my_df['rooms'] = ca_my_df['total_rooms'] / ca_my_df['households']
ca_my_df['bedrooms'] = ca_my_df['total_bedrooms'] / ca_my_df['households']

# 3. We can calculate the average of people in households
ca_my_df['people_per_house'] = ca_my_df['population'] / ca_my_df['households']
# the drop total_rooms, bedrooms, households, population
ca_my_df = ca_my_df.drop(['total_rooms','total_bedrooms','population','households'],axis=1)

ca_my_df.head(5)
```

```
def getAPISet(api, gram = 1):
    _set = set()
    for i in range(len(api) - gram + 1):
        key = ''
        for g in range(gram):
            key += api[i+g] # api
        _set.add(key)
    return _set 
```

```
APISet = set()
for key in data:
    APISet = APISet.union(getAPISet(data[key]))
APISet_li = list(APISet)
APISet_li
```

```
APIDict = {i:api for i,api in enumerate(APISet_li)}
APIDict.update({api:i for i, api in enumerate(APISet_li)})
APIDict
```

```
onehot = []

for key in data:
    onehot_api_li = [0]*19
    for api in data[key]:
        onehot_api_li[APIDict[api]] = 1
    onehot.append(onehot_api_li)
```

```
key_li = []
for key in data:
    key_li.append(key)
```

```
df = pd.DataFrame(onehot, columns=APISet_li, index=key_li)
df.sum().sort_values(ascending=False)
```

```
# down dimension by PCA and choose 2 princial compents to cluster
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA

X = np.array(onehot)

pca = PCA(n_components=2)
X_r = pca.fit(X).transform(X)

print("explained variance ratio: %s" % str(pca.explained_variance_ratio_))

plt.figure()
plt.scatter(X_r[:, 0], X_r[:, 1], alpha=.2)
plt.show()
```
